```python
import matplotlib.pyplot as plt
import numpy as np
years=['2000','2005','2010','2025','2020' ]
sachin=[300,600,1400,1200,1000] 
kohli=[100,1200,800,1500,800] 
sehwag=[0,200,400,1600,300]
x=np.arange(len(years))
width=0.2
plt.bar(x-width,sachin,width=0.2,label="Sachin",color='purple',edgecolor='black')
plt.bar(x,kohli,width=0.2,label="Sachin",color='pink',edgecolor='black')
plt.bar(x+width,sehwag,width=0.2,label="Sehwag",color='red',edgecolor='black')
plt.xticks(x,years)
plt.xlabel("Years")
plt.ylabel("Runs scored")
plt.title("Comparision of Runs scored by kohli,sachin and sehwag")
plt.legend(loc='upper left')
plt.grid(linestyle='--')
plt.tight_layout() 
```

```python
players = ['Sachin', 'Kohli', 'Sehwag']
runs = [15000, 12000, 9000]
plt.barh(players, runs,height=0.3, color='#BF8DF0',edgecolor='black')
plt.xlabel('Runs')
plt.ylabel("players")
plt.title('Total Runs Scored by Players')
plt.grid(linestyle='--')
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 500) #4pi means 2 cycle
x_sin = np.sin(x)
x_cos = np.cos(x)
x_exp = np.exp(x/ (4 * np.pi)) #np.exp(x / (4 * np.pi)) used bcoz this scales down the x so the exponential grows more slowly.  
plt.plot(x, x_sin,  label='sin(x)', color='blue',  linestyle='-')  
plt.plot(x, x_cos,  label='cos(x)', color='green', linestyle='--')  
plt.plot(x, x_exp,  label='exp(x)', color='red',   linestyle='-.')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('sin, cos, exp')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
